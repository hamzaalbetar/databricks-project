```python
df = spark.read.format("delta").table("workspace.bronze.erp_cust_az12")
```

```python
from pyspark.sql.functions import col, when, substring, length

df = df.withColumn(
    "cid",
    when(col("cid").startswith("NAS"), substring(col("cid"), 4, length(col("cid"))))
    .otherwise(col("cid"))
)
```

```python
from pyspark.sql.functions import col, current_date, when

df = df.withColumn(
    "bdate",
    when(col("bdate") > current_date(), None).otherwise(col("bdate"))
)
```

```python
from pyspark.sql.functions import col, upper, trim, when

df = df.withColumn(
    "gen",
    when(upper(trim(col("gen"))).isin("F", "FEMALE"), "Female")
    .when(upper(trim(col("gen"))).isin("M", "MALE"), "Male")
    .otherwise("n/a")
)
```

```python
df.write.mode("overwrite").saveAsTable("silver.erp_cust_az12")
```
