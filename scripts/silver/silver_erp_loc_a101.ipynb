```python
df = spark.read.format("delta").table("workspace.bronze.erp_loc_a101")
```

```python
from pyspark.sql.functions import regexp_replace

df = df.withColumn("cid", regexp_replace("cid", "-", ""))
```

```python
from pyspark.sql.functions import trim, when, col

df = df.withColumn(
    "cntry",
    when(trim(col("cntry")) == "DE", "Germany")
    .when(trim(col("cntry")).isin("US", "USA"), "United States")
    .when(col("cntry").isNull() | (trim(col("cntry")) == ""), "n/a")
    .otherwise(trim(col("cntry")))
)
```

```python
df.write.mode("overwrite").saveAsTable("silver.erp_loc_a101")
```
