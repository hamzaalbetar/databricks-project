```python
df = spark.read.format("delta").table("workspace.bronze.crm_cust_info")
```

```python
from pyspark.sql.functions import trim

df = df.withColumn("cst_firstname", trim(df["cst_firstname"])) \
               .withColumn("cst_lastname", trim(df["cst_lastname"]))
```

```python
from pyspark.sql.functions import when, upper, col

df = df.withColumn(
    "cst_marital_status",
    when(upper(col("cst_marital_status")) == "S", "Single")
    .when(upper(col("cst_marital_status")) == "M", "Married")
    .otherwise(col("cst_marital_status"))
)
```

```python

df = df.withColumn(
    "cst_gndr",
    when(upper(col("cst_gndr")) == "M", "Male")
    .when(upper(col("cst_gndr")) == "F", "Female")
    .otherwise(col("cst_gndr"))
)
```

```python
df.write.mode("overwrite").saveAsTable("silver.crm_cust_info")
```
