```python
df = spark.sql("""
SELECT 
  ROW_NUMBER() OVER (ORDER BY ci.cst_id) AS customer_key,
  ci.cst_id AS customer_id,
  ci.cst_key AS customer_number,
  ci.cst_firstname AS first_name,
  ci.cst_lastname AS last_name,
  la.cntry AS contry,
  ci.cst_marital_status AS marital_status,
  CASE WHEN ci.cst_gndr != 'n/a' THEN ci.cst_gndr
       ELSE COALESCE(cg.gen, 'n/a')
  END AS gender,
  cg.bdate AS birthdate,
  ci.cst_create_date AS create_date
FROM workspace.silver.crm_cust_info AS ci
LEFT JOIN workspace.silver.erp_loc_a101 AS la
  ON ci.cst_key = la.cid
LEFT JOIN workspace.silver.erp_cust_az12 AS cg
  ON ci.cst_key = cg.cid
""")
```

```python
df.write.mode("overwrite").saveAsTable("gold.dim_customers")
```
