#reading file (crm_cust_info.csv)

```python
df=(
    spark.read.option("header","true")
    .option("inferSchema","true")
    .csv("/Volumes/workspace/default/base/datasets/source_crm/cust_info.csv")
)
```

#write (crm_cust_info.csv) into bronze layer

```python
df.write.mode("overwrite").saveAsTable("bronze.crm_cust_info")
```
